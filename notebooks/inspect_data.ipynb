```python
import csv

# Your data
steps = data[1]['steps'][:data[1]['first_essp_index']+1] # your full list

# Create CSV
with open('../data/reasoning_steps_problem2.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    
    # Header row
    writer.writerow(['step_number', 'step_text', 'behavioral_category', 'subcategory', 
                     'notes', 'is_handoff_sufficient', 'is_early_stop_sufficient', 
                     'confidence', 'tags'])
    
    # Data rows
    for i, step in enumerate(steps, start=1):
        is_handoff = 'TRUE' if i <= 38 else 'FALSE'
        is_early_stop = 'TRUE' if i <= 72 else 'FALSE'
        writer.writerow([i, step, '', '', '', is_handoff, is_early_stop, '', ''])

print("CSV created! Upload reasoning_steps.csv to Google Sheets")
```

```python
import json
data = []
with open('../data/hsp_step_accuracies.jsonl', 'r') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
import matplotlib.pyplot as plt

for j in range(len(data)):
    
    plt.figure(figsize=(7,4))
    plt.plot(
        [i+1 for i, _ in enumerate(data[j]['essp_step_accuracies'])],
        data[j]['essp_step_accuracies'],
        color="blue"
    )
    plt.title('ESSP Accuracies')
    plt.show()
    plt.close()
    
    plt.figure(figsize=(7,4))
    plt.plot(
        [i+1 for i, _ in enumerate(data[j]['hsp_step_accuracies'])],
        data[j]['hsp_step_accuracies'],
        color="red"
    )
    plt.title('HSP Accuracies')
    plt.show()
    plt.close()
```
