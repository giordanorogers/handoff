```python
import json

data = []
with open ('../data/MATH_92/MATH_92.jsonl', 'r') as f:
    for line in f:
        data.append(json.loads(line))
len(data)
```

```python
data[0].keys()
```

```python
for i, row in enumerate(data):
    row['id'] = i+1
```

```python
data[0].keys()
```

```python
with open ('../data/MATH_92/MATH_92_indexed.jsonl', 'w') as f:
    for item in data:
        f.write(json.dumps(item) + '\n')
```
